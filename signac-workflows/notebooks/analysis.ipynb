```python
import matplotlib.pyplot as plt
import numpy as np
import signac
```

```python
project = signac.get_project("../test_ensembles/")
```

```python
for job in project.find_jobs({"ensemble": "NVE", "displacement":2.5}):
    data = np.genfromtxt(job.fn("log-shear.txt"), names=True)
    temp = data["mdcomputeThermodynamicQuantitieskinetic_temperature"]
    plt.plot(temp, label="Shear", linestyle="-", markersize=0)
    print(np.mean(temp[-2000:]))

    data = np.genfromtxt(job.fn("log-tensile.txt"), names=True)
    temp = data["mdcomputeThermodynamicQuantitieskinetic_temperature"]
    plt.plot(temp, label="Tensile")
    print(np.mean(temp[-2000:]))

plt.legend()
```
